```python
import numpy as np

# Custom activation functions
def sigmoid_activation(val):
    return 1.0 / (1.0 + np.exp(-val))

def tanh_activation(val):
    return np.tanh(val)

# All given weights and biases grouped together
lstm_weights = {
    'Wf': 0.5, 'Whf': 0.1, 'bf': 0.0,
    'Wi': 0.6, 'Whi': 0.2, 'bi': 0.0,
    'Wc': 0.7, 'Whc': 0.3, 'bc': 0.0,
    'Wo': 0.8, 'Who': 0.4, 'bo': 0.0
}

def process_sequence(seq):
    # Initialize states to zero
    h_state = 0.0
    c_state = 0.0

    print("=== Starting Manual LSTM Computation ===")

    for step, current_x in enumerate(seq):
        t = step + 1
        if t == 4:
            print(f"\n[ Time Step {t} | Predicting next value using input = {current_x} ]")
        else:
            print(f"\n[ Time Step {t} | Input x = {current_x} ]")

        # 1. Calculate Forget Gate
        f_raw = lstm_weights['Wf'] * current_x + lstm_weights['Whf'] * h_state + lstm_weights['bf']
        f_gate = sigmoid_activation(f_raw)
        print(f" -> Forget Gate (f): {f_gate:.3f}")

        # 2. Calculate Input Gate
        i_raw = lstm_weights['Wi'] * current_x + lstm_weights['Whi'] * h_state + lstm_weights['bi']
        i_gate = sigmoid_activation(i_raw)
        print(f" -> Input Gate (i): {i_gate:.3f}")

        # 3. Calculate Candidate Cell State
        c_cand_raw = lstm_weights['Wc'] * current_x + lstm_weights['Whc'] * h_state + lstm_weights['bc']
        c_cand = tanh_activation(c_cand_raw)
        print(f" -> Candidate State (C~): {c_cand:.3f}")

        # 4. Update Cell State
        c_state = (f_gate * c_state) + (i_gate * c_cand)
        print(f" -> Updated Cell State (C): {c_state:.3f}")

        # 5. Calculate Output Gate
        o_raw = lstm_weights['Wo'] * current_x + lstm_weights['Who'] * h_state + lstm_weights['bo']
        o_gate = sigmoid_activation(o_raw)
        print(f" -> Output Gate (o): {o_gate:.3f}")

        # 6. Update Hidden State
        h_state = o_gate * tanh_activation(c_state)
        print(f" -> Updated Hidden State (h): {h_state:.3f}")

    # Final linear transformation prediction
    print("\n=== Prediction Phase ===")
    W_y = 4.0
    b_y = 0.0
    predicted_value = (W_y * h_state) + b_y

    print(f"Predicted Output = {W_y} * {h_state:.3f} + {b_y} = {predicted_value:.3f}")
    print("The result is close to 4 as expected.")

if __name__ == '__main__':
    input_data = [1, 2, 3, 4]
    process_sequence(input_data)
```
